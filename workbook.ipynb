```python
import pandas as pd
```

```python
df = pd.read_excel(r'excel\Axel LYO Vials.xlsx')
```

```python
df = df.loc[df['STORE_CODE'] == 'PSLYO']
```

```python
# Transforming datas into correct types
df.loc[:,'DATE_REGISTERED'] = pd.to_datetime(df['DATE_REGISTERED'])
df['VIAL_BARCODE'] = df['VIAL_BARCODE'].astype(str)
```

```python
solutions = [' lsg', ' lösung', ' losung', ' loesung', ' solut', ' sol.', ' soln', ' THF', ' ethyl acetate', ' dmf', ' dimethylformamide'
            ' dioxane', ' tetrahydrofuran',' diethylether', ' diethyl ether', ' MTBE', ' methyltertbutylether', ' ethyl acetate', ' toluene', ' toluol', ' aceton', 
            ' dcm', ' dichloromethan', ' methylenchlorid', 'WT%', 'WT %']
exclusion_solutions = ['anisol']

hydride = ['hydrid']   
exclusion_hydride = ['anhydrid']



refrigeration = ['cooling', 'freezer']
risk = ['1', '4.2', '4.3', '5.1A', '5.2']
cas_number = ['302-01-2', '10025-87-3', '75-35-4', '75-77-4', '1113-78-6', '10294-34-5', '10294-33-4', '75-52-5', '18279-73-7', '625-36-5', '14337-43-0'] 
#Hydrazine, Oxalyl chloride, 1,1-dichloroethylene, TMS-Cl, TRI-SEC-Butylborane, Boron trichloride, BBr3, Nitromethan, chloropropionyl chloride,
# => résultats sécurité depuis 2021

str_id = [1617581, 456227, 10133, 1884, 1268, 2843, 3992, 4578, 4755, 456227, 684291, 904810]
```

```python
#Preparing filters
filter_name_sol = df['Product Name'].str.contains('|'.join(solutions), na=False, case=False)
filter_name_hydride = df['Product Name'].str.contains('|'.join(hydride), na=False, case=False)

filter_exclusion_solutions = df['Product Name'].str.contains('|'.join(exclusion_solutions), na=False, case=False)
filter_exclusion_hydride = df['Product Name'].str.contains('|'.join(exclusion_hydride), na=False, case=False)



filter_refrigeration = df['Refrigeration'].str.contains('|'.join(refrigeration), case=False, na=False)
filter_storageclass = df['Storage Class'] == '|'.join(risk)
filter_cas = df['CAS Number'].str.contains('|'.join(cas_number), na=False, case=False)
filter_id = df['Str. ID'].isin(str_id)
```

```python
df_filtered = df[(filter_name_sol&~filter_exclusion_solutions)
                 |(filter_name_hydride&~filter_exclusion_hydride)
                 |filter_refrigeration
                 |filter_storageclass
                 |filter_cas
                 |filter_id]
```
